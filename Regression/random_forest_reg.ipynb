# Random forest regression
**Steps:**
1. Pick at random 'K' data points from the Training set.
2. Build the Decision Tree associated to these 'K' data points.
3. Choose the number N-tree of trees you want to build & repeat STEPS 1 & 2.
4. While Predicting for a new data point, Each tree predicts for the specific data point, Average of all predictions is the final predicted value.
- Every decision tree has high variance, but when we combine all of them in parallel then the resultant variance is low as each decision tree gets perfectly trained on that particular sample data, and hence the output doesnâ€™t depend on one decision tree but on multiple decision trees
### Note:
- Random forest is an example of **Ensemble learning**, when you take multiple algorithms or the same algorithem multiple times and you put them together that results in a much more powerfull version.

### Importing the dataset

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

dataset = pd.read_csv('./reg_dataset/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
Y = dataset.iloc[:, -1].values

Y = np.reshape(Y,(len(Y),1))
```

### Training the dataset

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X, Y)
regressor.predict([[6.5]])
```

```python
plt.title("Salary vs Levels")
plt.xlabel("position levels")
plt.ylabel("Salary")
plt.scatter(X,Y, color="red")

x_grid = np.arange(min(X), max(X), 0.1)
x_vals = np.reshape(x_grid,(len(x_grid),1))
y_vals = regressor.predict(x_vals)
plt.plot(x_vals,y_vals, color='blue')

plt.show()
```
