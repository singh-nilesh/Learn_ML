# Eclat 
**is an association rule leaning that analyzes that people who bought product set A also bought product set B.**

-  Equivalence Class Clustering and Bottom-Up Lattice

-  more efficient and scalable version of the Apriori algorithm.

- Apriori &rarr; Horizontal approach, Breadth-first-search
- Eclat  &rarr; Veartical approach, Depth-First-search ( **Faster then Apriori** )

- Only based on **support**, of set of products, (more than one products)
    - Support refers to the default popularity of any product.
    - $$ \textrm{support(set A)} = \frac {\textrm{transaction  containing set A}}{\textrm{Total transaction}}$$

### Steps :
1. Set a minimum Support
1. Take all the subsets in transactions having higher support than minimum support
1. Sort these subsets by decreasing support

### Implementing Eclat model



```python
# Importing libaraies
import pandas as pd
import numpy as np

# Importing dataset
dataset = pd.read_csv("Market_Basket_Optimisation.csv", header = None)
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])

# Training Eclat
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)



# Visualising results
results = list(rules)

def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(lhs, rhs, supports))

resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['LHS', 'RHS', 'Support'])
resultsinDataFrame.nlargest(n = 10, columns = 'Support') # Sorts the values in the dataframe in descending order as per Lift
```

- Although, Eclat is faster than Apriori, Apriori is more refined and favored algorithem, due to is more reliable and usefull results.
