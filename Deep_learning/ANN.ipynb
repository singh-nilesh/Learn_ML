### The Activation Function.

- decides whether a neuron should be activated or not, In other words, it decides importance of its Signal passed to other Nodes.
- It introduces non-linearity into the output of a neuron.
- Achived by calculating the weighted sum and further adding bias to it.
- **Bias :** the constant which is added to the product of features and weights. It is used to offset the result. 
- It helps the models to shift the activation function towards the positive or negative side, Favor or Exclude nodes.

    $$ f(b+\sum_{i=1}^n x_iw_i) = [w_1a_1 + w_2a_2 + w_2a_3 + ... + w_na_n + (-10)bias]$$
### Note
-  Weight increases the steepness of activation function. This means weight decide how fast the activation function will trigger 
- Whereas bias is used to delay the triggering of the activation function. 

### Types of Activation function

    - it is especially used for models where we have to predict the probability as an output.
    - When the output value is close to 1, the neuron is active and enables the flow of information, while a value close to 0 corresponds to an inactive neuron.

### How do Neural Networks Learn ?
- When we refear '**Neural Network is leraning**' it  is just Algorithem trying to **Minimize the Cost function** of prediction with help of **Back Propogation** algoritem.

**Phase 1**
1. Multiple inputs are fead to Model.

1. The hidden layer consists of various neurons(nodes) & layers of nodes.
    - Each layer of nodes focus on diffrent aspects of the Input data, and consider various condition/stucture to recreate the Input.
    - Diffrent **Activation functions** & **Weights of Synapses** are used to make decisions and control the flow of Algorithem.
        <img src='./deepLrn_photos/NN_phase1.png'>

1. Based on Conditions and finding, Output is generated.
    - There can be **multiple Output generated**, this results in Uncertainity, and Unreliable outcome.

**Phase 2**
1. When the model generates **Multiple unreliable outputs**, In needs to be given Feedback, So that it can improve. So A **Cost Function** is used.

1. **Cost function :**  It is measures of error between the predicted and actual values and helps the network to adjust its weights and biases to make accurate predictions.
    - The goal is to find the minimum value of the cost function, which corresponds to the best set of weights and biases that make accurate predictions.

        <img src='./deepLrn_photos/cost_fun.png' height='300px'>

1. Based on the Cost function The Weights and Bias in the hidden layer are re-adjusted in a reverse manner known as **BackPropagation**. to minimuze to cost and attaing desired result.

1. **Back Propagation :**  an **iterative** algorithm, that helps to minimize the cost function by determining which weights and biases should be adjusted. During **every epoch**.
    - All the weights at same time.
    - the model learns by **adapting the weights and biases** to minimize the loss by moving down toward the gradient of the error. 
    - Algorithems Used in Back prooppogation are 
        1. **Batch Gradient Descent**,when Cost function is calculated by taking average of all outcomes, finally weights are adjusted.

        1. **Stochastic Gradient Descent :**, when Cost function is calculated in **every Round (each Epoch).**, weights are adjusted in Every round.

**Repate** Phase 1 & Phase 2.

**This Cycle of Producing output and Re-factoring weights and bais until desired results is obtained. is called as 'NN learnig.'**

### Minimum using Stochastic Gradient Descent (SGD)
- Instead of using the entire dataset for each iteration, only a single random training example (or a small batch) is selected to calculate the gradient and update the model parameters.
- This random selection introduces randomness into the optimization process, hence the term “stochastic” in stochastic Gradient Descent

### Steps for Implementation :
1. Randomly initialize the weights to small number close to 0(but not 0).
2. Input the first observation of your dataset in the input layer, each feature in one input node.
3. Forward-Propagation: from left to right, the neurons are activated in a way that impact of each neuron's activation is limited by the weights. Propagate the activations until getting the predicted result y.
4. Compare the predicted result to the actual result. Measure the generated error.
5. Back-Propagation: from right to left, the error is back-Propogated. Update the weights according to how much they are responsible for the error. The learninig rate decides by how much we update the weights.

6. Repeat Steps 1 to 5 and update the weights after each observation (Reinforcement learning) Or :  Steps 1 to 5 and update the weights only after a batch of observations (Batch Learning).

### Pre-processing

```python
 # Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf

tf.__version__
```

```python
# IMporting the dataset
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values

print('Independent var ---> ',X)
print('Dependent var --->',y)
```

```python
# Encoding categorical data, transform Gender column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])

# Performing one hot encoding for Column(2) ---> Country
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))

print(X[0])
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print(X_train[0])
```

**Note:** Feature Scalling is very very Crucial for Deep Learning.

# Building ANN

```python
# initializing the ANN
ann = tf.keras.models.Sequential()

# Adding the input layer and the first hidden layer
# there is no specific way to decide the number of neurons, but can be achived by experimenting with diffrent Hyper-parameters.
# in a fully connected NN, activation = Rectifier
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))

# Adding second Hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))


# Adding the output Layer
# A fully-connecter(Dense) layer to hidden layer, with 1 output node(Binary)(units), activation = Sigmoid
# We are using sigmoid because, of Binary output, Incase of Non-binary output, activation='softmax'
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

The model would look like this, where Every node is connected to Every other node in Layer Adjcent to it.

<img src='./deepLrn_photos/ann_model_str.png'>

### Training the ANN

1. **Adam Optimizer:** Adam computes individual adaptive learning rates for different parameters. It adapts the learning rate for each parameter by storing the moving average of the gradients and the squared gradients.
    - Similar to **Stochastic gradient descent**.

```python
# We will be using Stochastic Gradient Descent optimizer = 'adam'
# this case is Binary Classification i.e loss='binary_crossentropy'
# In case of Non-Binary Classification( 3 or more Outputs ),  loss='categorical_crossentropy'
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Training the model
ann.fit(X_train, y_train, batch_size=32, epochs=100)
```

### Predicting Results

```python
# p1= [['France'=[1,0,0], 600, 'Male'=1, 40, 3, 60000, 2, 1, 1, 50000]]
p1= [[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]]
probability = ann.predict(sc.transform(p1))

if probability > 0.5 : 
    print(f'Will not leave the company ---> {round(probability[0][0]*100,2)}%')
else:
    print(f'Will leave the company ---> {round(probability[0][0]*100,2)}%')
```

The values of prediction for diffrent ANN models (Every time you re-train) will vary, So consider an Average, Or Fine tune the model as per your dataset. 

### Predicting and Visualizing test results

```python
# predicting test set
y_pred = ann.predict(X_test)

# transforming y_pred, ---> 0 if p(A) < 50% else, 1
y_pred = [[0] if x[0]<0.5 else [1] for x in y_pred]

# makinng the confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_pred,y_test)
print(cm)
print(f" \t Accuracy ---> {accuracy_score(y_pred,y_test)}")
```
